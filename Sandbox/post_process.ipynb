```python
import pandas as pd
import numpy as np
import shutil
import os
import time
import subprocess
import sys
import datetime
import matplotlib.pyplot as plt
from os.path import exists as file_exists
import matplotlib.dates as mdates


study = ["baseline","outage"]
og_wd = os.getcwd()
data_file_path  =  str(og_wd)
data_file_path = data_file_path.replace(os.sep, '/')
outputs_path = data_file_path + '/Outputs/'
models_path = data_file_path + '/Models/'

# Identify all the models
models = []
# Iterate directory
for model in os.listdir(models_path):
    # check only text files
    if model.endswith('.osm'):
        models.append(model[:-4])
        
# Outage 
otg_date = "July 24"
otg_hour = "12"
otg_lengths = os.listdir(outputs_path + "outage")
otg_lengths.sort(key=int)

master_baseline = pd.DataFrame()
```

```python
# Plots 
fig_num = 1

for each_model in models:
    
    data = [['07/24 00:00:00'],['07/26 23:00:00']]
    index = ['start','end']
    columns = ['Date/Time']
    outage = pd.DataFrame(data,index,columns)
    outage['Date/Time']= pd.to_datetime(outage['Date/Time'], format = '%m/%d %H:%M:%S')

    ## Baseline
    baseline = pd.DataFrame()
    data = pd.read_csv(outputs_path + 'baseline/' + each_model + '/' + each_model + '_baseline_eplusout.csv')
    data = data.loc[data['LIVING ZONE:Zone Air Temperature [C](Hourly)'].notnull()]
    baseline['Date/Time'] = data['Date/Time']
    baseline['Outdoor Temp'] = data['LIVING ZONE:Zone Outdoor Air Drybulb Temperature [C](Hourly)']
    baseline[each_model] = data['LIVING ZONE:Zone Air Temperature [C](Hourly)']
    baseline['Electricity'] = data['Electricity:Facility [J](Hourly)']
    baseline['Gas'] = data['NaturalGas:Facility [J](TimeStep) ']

    # Change '24:00:00' to "00:00:00", set datetime, add one day to midnight values, and reset index
    baseline['Date/Time'] = baseline['Date/Time'].str.replace('24:00:00','00:00:00')
    baseline['Date/Time'] = pd.to_datetime(baseline['Date/Time'],format = ' %m/%d  %H:%M:%S')
    for eachIndex in baseline.index:
        if baseline['Date/Time'][eachIndex].hour == 0:
            baseline['Date/Time'][eachIndex] = np.datetime64(baseline['Date/Time'][eachIndex]) + np.timedelta64(1,'D')

    baseline['Date/Time'] = pd.to_datetime(baseline['Date/Time'],format = ' %m/%d  %H:%M:%S')

    baseline = baseline.reset_index(drop=True)
    baseline = baseline.loc[(baseline['Date/Time'] >= outage['Date/Time'][0]) & (baseline['Date/Time'] <= outage['Date/Time'][1])]
    baseline = baseline.reset_index(drop=True)

    #plot baseline
    plt.figure(fig_num)
    fig_num = fig_num + 1
    plt.plot(baseline['Date/Time'],baseline['Outdoor Temp'], label = 'Outdoor Temp')
    plt.plot(baseline['Date/Time'],baseline[each_model], label = 'Indoor Temp')
    plt.legend()
    plt.title('Baseline for ' + each_model)
    plt.xticks(rotation=90)
    plt.ylim(18,34)
    plt.xlabel('Date')
    plt.ylabel('Temperature (C)')
    save_path = data_file_path + '/Figures/' + each_model + '_baseline_plot.png'
    plt.savefig(save_path)
    plt.show()
    
    master_baseline[each_model + "_Tin"] = baseline[each_model]
    master_baseline[each_model + "_E"] = baseline["Electricity"]
    master_baseline[each_model + "_G"] = baseline["Gas"]

    
    if each_model == models[0]:
        master_baseline['Date/Time'] = baseline['Date/Time']
        master_baseline['Outdoor Temp'] = baseline['Outdoor Temp']
    for each_otg in otg_lengths:
        master_outage = pd.DataFrame()
        master_diff = pd.DataFrame()
        if each_model == models[0]:
            master_outage['Date/Time'] = baseline['Date/Time']
            master_outage['Outdoor Temp'] = baseline['Outdoor Temp']
            master_diff['Date/Time'] = baseline['Date/Time']
            master_diff['Outdoor Temp'] = baseline['Outdoor Temp']
                
        # Get Date Times
        otg_start = pd.to_datetime(otg_date + " " + otg_hour + ":00:00",format = "%B %d %H:%M:%S")
        otg_hr_end = int(otg_hour) + int(each_otg)
        
        if otg_hr_end < 24:
            otg_end = pd.to_datetime(otg_date + " " + str(otg_hr_end) + ":00:00",format = "%B %d %H:%M:%S")
        elif otg_hr_end == 24:
            otg_end = pd.to_datetime(otg_date + " 00:00:00", format = "%B %d %H:%M:%S")
            otg_end = otg_end + datetime.timedelta(days=1)
            otg_hr_end = otg_end.hour 
        else: 
            otg_end = pd.to_datetime(otg_date + " 00:00:00", format = "%B %d %H:%M:%S")
            otg_end = otg_end + datetime.timedelta(days=1) 
            otg_end = otg_end + datetime.timedelta(hours = (otg_hr_end - 24))
            otg_hr_end = otg_end.hour
        outage_comparison = pd.DataFrame()
                
        data = pd.read_csv(data_file_path + '/Outputs/outage/' + each_otg + "/" + each_model + '/' + each_model + '_outage_' + each_otg + '_eplusout.csv')
        data = data.loc[data['LIVING ZONE:Zone Air Temperature [C](Hourly)'].notnull()]
        outage_comparison['Date/Time'] = data['Date/Time']
        outage_comparison['Outdoor Temp'] = data['LIVING ZONE:Zone Outdoor Air Drybulb Temperature [C](Hourly)']
        outage_comparison[each_model] = data['LIVING ZONE:Zone Air Temperature [C](Hourly)']
        outage_comparison['Electricity'] = data['Electricity:Facility [J](Hourly)']
        outage_comparison['Gas'] = data['NaturalGas:Facility [J](TimeStep) ']

        outage_comparison['Date/Time'] = outage_comparison['Date/Time'].str.replace('24:00:00','00:00:00')
        outage_comparison['Date/Time'] = pd.to_datetime(outage_comparison['Date/Time'],format = ' %m/%d  %H:%M:%S')

        for eachIndex in outage_comparison.index:
            if outage_comparison['Date/Time'][eachIndex].hour == 0:
                outage_comparison['Date/Time'][eachIndex] = np.datetime64(outage_comparison['Date/Time'][eachIndex]) + np.timedelta64(1,'D')

        outage_comparison=outage_comparison.reset_index(drop=True)

        # Reduce to outage date
        outage_comparison = outage_comparison.loc[(outage_comparison['Date/Time'] >= outage['Date/Time'][0]) & (outage_comparison['Date/Time'] <= outage['Date/Time'][1])]
        outage_comparison = outage_comparison.reset_index(drop=True)

        # plot outage
        plt.figure(fig_num)
        fig_num = fig_num + 1
        plt.plot(outage_comparison['Date/Time'],outage_comparison['Outdoor Temp'], label = 'Outdoor T')
        plt.plot(outage_comparison['Date/Time'],outage_comparison[each_model], label = 'Outage Indoor T')
        plt.plot(baseline['Date/Time'],baseline[each_model], label = 'Baseline Indoor T')       
        plt.title(each_otg + ' Hour Power Outage Comparison')
        plt.xlabel('Date')
        plt.ylabel('Temperature (C)')
        plt.ylim(18,34)
        plt.xticks(rotation=90)
        plt.axvspan(otg_start, otg_end, ymin=0, ymax=34, alpha=0.25, color='gray',label = 'Power Outage')
        plt.legend()
        save_path = data_file_path + '/Figures/' + each_model + '_' + each_otg + '_outage_and_baseline_plot.png'
        plt.savefig(save_path)

        delta = pd.DataFrame()
        delta['Date/Time'] = baseline['Date/Time']
        delta['Outdoor Temp'] = baseline['Outdoor Temp']
        delta['Baseline'] = baseline[each_model]
        delta['Diff'] = outage_comparison[each_model] - baseline[each_model]

        #reduce delta just to outage
        delta = delta.loc[(delta['Date/Time'] >= outage['Date/Time'][0]) & (delta['Date/Time'] <= outage['Date/Time'][1])]

        plt.figure(fig_num)
        fig_num = fig_num + 1
        plt.plot(delta['Date/Time'],delta['Diff'], label = 'Temp Diff')
        plt.title('Difference from Baseline During ' + each_otg + ' Hour Power Outage for ' + each_model)
        plt.xticks(rotation=90)
        plt.ylim(0,5)
        plt.axvspan(otg_start, otg_end, ymin=0, ymax=5, alpha=0.25, color='gray',label = 'Power Outage')
        plt.legend()
        plt.xlabel('Date')
        plt.ylabel('Temperature (C)')
        save_path = data_file_path + '/Figures/' + each_model + "_" + each_otg +'_outage_temp_plot.png'
        plt.savefig(save_path)

        # plot electricity difference
        plt.figure(fig_num)
        fig_num = fig_num + 1
        plt.plot(outage_comparison['Date/Time'],outage_comparison['Electricity']/1000000, label = 'Outage')
        plt.plot(baseline['Date/Time'],baseline['Electricity']/1000000, label = 'Baseline')
        plt.axvspan(otg_start, otg_end, ymin=0, ymax=8, alpha=0.25, color='gray',label = 'Power Outage')
        plt.title('Electricity Use During ' + each_otg + ' Hour Power Outage Comparison')
        plt.xticks(rotation=90)
        plt.ylim(0,8)
        plt.xlabel('Date')
        plt.ylabel('Electricity [ J 1e6]')
        plt.legend()
        save_path = data_file_path + '/Figures/' +  each_model + '_' + each_otg + '_electricity_plot.png'
        plt.savefig(save_path)
        
        master_outage[each_model + "_Tin"] = outage_comparison[each_model]
        master_outage[each_model + "_E"] = outage_comparison["Electricity"]
        master_outage[each_model + "_G"] = outage_comparison["Gas"]
    
        master_diff[each_model] = delta['Diff']
    master_diff.to_csv(data_file_path + "/Figures/Data/master_diff_" + each_otg + ".csv",index=False)
    master_outage.to_csv(data_file_path + "/Figures/Data/master_outage_" + each_otg + ".csv",index=False)
master_baseline.to_csv(data_file_path + "/Figures/Data/master_baseline.csv",index=False)
```
