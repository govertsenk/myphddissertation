```python
## THIS FILE IS IMPORTANT BECAUSE READVARSESO.EXE CANNOT BE RUN ON DISCOVERY
## SO YOU MUST DOWNLOAD YOUR FOLDERS/FILES TO YOUR WINDOWS COMPUTER
## SAVE READVARSESO.EXE IN YOU FOLDER
## RUN THIS NOTEBOOK TO EXECUTE THE PROGRAM 
## AND MOVE ALL OF YOUR eplusout.csv FILES INTO ONE FOLDER SO YOU CAN THEN
## REUPLOAD THE OUTPUTS TO DISCOVERY FOR FURTHER ANALYSIS

# IMPORTS 
import pandas as pd
import numpy as np
import shutil
import os
import time
import subprocess
import sys
import datetime
import matplotlib.pyplot as plt
from os.path import exists as file_exists
import matplotlib.dates as mdates
import math
import matplotlib.ticker as mtick

# CURRENT DIRECTORY
og_wd = os.getcwd()

# FILE PATH  -- EDIT AS NECESSARY
data_file_path = og_wd + "/Outputs/trial_1"
data_file_path = data_file_path.replace(os.sep, '/')

# CONSTANTS -- EDIT AS NECESSARY
study = ["baseline"]

# Identify all the models
models = os.listdir(data_file_path + "/" + study[0])
# Make new Outputs folders
new_path = data_file_path + "/" + "output"
if not os.path.exists(new_path):
     os.makedirs(new_path)
                      
for each_study in study:
    new_path = data_file_path + "/" + "output" + "/" + each_study 
    if not os.path.exists(new_path):
        os.makedirs(new_path)
```

```python
for each_model in models:
    # Name of the program
    program_name = "/ReadVarsESO.exe"
    
    # Path to program. you can change this if needed
    source = data_file_path + program_name
    for each_study in study:
        # Copy and Paste Read Vars ESO
        new_path = data_file_path + "/" + each_study + "/" + each_model + "/"
        destination = new_path + program_name
        shutil.copyfile(source,destination)

        # Change to That folder
        os.chdir(os.path.abspath(new_path))

        # Run ReadVarsEso.exe
        subprocess.run(["ReadVarsESO"], shell=True)

        # Copy and Paste New eplusout CSV to output folder with name of model
        old_location = new_path + "eplusout.csv"
        new_location = data_file_path + "/" + "output" + "/" + each_study + "/" + each_model + ".csv"
        shutil.copyfile(old_location,new_location)

        # Delete ReadVarsEso.exe from the folder
        os.remove(destination)

        # Change Back to Original Directory
        os.chdir(og_wd)
```
